```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertia = []
K = range(1, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)',
                hue='Cluster', data=df, palette='viridis', s=60)
plt.title('Customer Segments based on Income and Spending Score')
plt.show()
```

```python
cluster_summary = df.groupby('Cluster')[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean()
print(cluster_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', 
                hue='Cluster', data=df, palette='Set2', s=60)
plt.title('Customer Segments by Income and Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1100)')
plt.legend(title='Cluster')
plt.show()
```

```python
### Task 02  Customer Segmentation Using K-Means Clustering

**Objective:**  
To identify distinct customer groups in a retail dataset using unsupervised learning.

**Algorithm:**  
K-Means Clustering

**Implementation Steps:**
1. Preprocessed and scaled data using StandardScaler.
2. Determined optimal number of clusters (K = 5) using Elbow Method.
3. Applied K-Means Clustering on 'Annual Income' and 'Spending Score'.
4. Visualized clusters and analyzed their demographic profiles.

**Results:**
- Formed **five unique customer segments** based on income and spending patterns.
- Key segment profiles:
  - Cluster 1: Premium / High-Spending customers  
  - Cluster 2: Young impulsive spenders  
  - Cluster 3: Wealthy but low-spending group  
  - Cluster 4: Low-income, low-spending budget customers  
  - Cluster 0: Balanced, average-spending customers

**Insights:**
- The business can target **Cluster 1** for loyalty programs and **Cluster 2** for offers and upselling.  
- **Cluster 3** may need trust-based marketing, while **Cluster 4** can be approached with budget-friendly deals.

**Tools Used:** Python, Pandas, NumPy, Scikit-Learn, Matplotlib, Seaborn
```
